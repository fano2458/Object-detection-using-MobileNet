```python
import cv2
import numpy as np
```

```python
image_path = "data/images/000009.jpg"
grid_size = 5
```

```python
image = cv2.imread(image_path)
```

```python
height, width = image.shape[:2]

# Calculate grid dimensions
grid_width = width // grid_size
grid_height = height // grid_size

# Create a new image with the same dimensions as the input image
grid_image = np.zeros_like(image)

# Draw vertical lines
for x in range(0, width, grid_width):
    cv2.line(grid_image, (x, 0), (x, height), (255, 0, 255), 1)

# Draw horizontal lines
for y in range(0, height, grid_height):
    cv2.line(grid_image, (0, y), (width, y), (255, 0, 255), 1)

# Combine the original image with the grid image
result = cv2.addWeighted(image, 0.7, grid_image, 0.3, 0)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(result)
```
